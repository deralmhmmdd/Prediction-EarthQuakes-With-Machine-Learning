```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

```python
df= pd.read_csv("Data KP Gempa Bumi Wilayah Sumatera Selatan Dan Sekitarnya.csv")
df
```

```python
X = df.iloc[:,:]
```

```python
y = df.iloc[:, 7]
```

```python
df['Time']
```

```python
minutes = []
hours = []
seconds = []

for t in df.Time:
    t  = t.split(":")
    minutes.append(t[1])
    hours.append(t[0])
    seconds.append(t[2])
```

```python
df['Time'] = pd.to_datetime(df['Time'])
```

```python
df['Jam'] = df['Time'].dt.hour
df['Menit'] = df['Time'].dt.minute
df['Detik'] = df['Time'].dt.second
```

```python
df.drop(['Time'], axis=1, inplace=True)
df.head()
```

```python
Places =pd.get_dummies(X['Place'],drop_first=True)
Places.head()
```

```python
X =X.drop('Place',axis=1)
```

```python
X = pd.concat([X,Places],axis =1)
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.2, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
df =pd.DataFrame({'Actual': y_test,'Predicted':y_pred})
df
```

```python
from sklearn import metrics  
print('Model R^2 Square value',metrics.r2_score(y_test,y_pred))
```
